```python
class Node:
  def __init__(self, e, n):
    self.element = e
    self.next = n
```

```python
class LinkedList:
  
  def __init__(self, a):
  #  Design the constructor based on data type of a. If 'a' is built in python list then
  #  Creates a linked list using the values from the given array. head will refer
  #  to the Node that contains the element from a[0]
  #  Else Sets the value of head. head will refer
  # to the given LinkedList

  # Hint: Use the type() function to determine the data type of a
    # To Do
    if type(a) == list:
        self.head = None
        self.tail = None
        for i in range(len(a)):
            n = Node(a[i], None)
            if self.head == None:
                self.head = n
                self.tail = n
            else:
                self.tail.next = n
                self.tail = n
    
    else:
        self.head = a
    
  
  # Count the number of nodes in the list
  def countNode(self):
    # To Do
    count = 0
    n = self.head
    while n != None:
        count += 1
        n = n.next
    return count
  
  # Print elements in the list
  def printList(self):
    # To Do
    n = self.head
    while n != None:
        print(n.element, end = "")
        n = n.next
        if n != None:
            print(",", end="")
        else:
            print()

  # returns the reference of the Node at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    size = self.countNode()
    if idx < 0 or idx >= size:
      return None
      
    else:
      count = 0
      temp = self.head

      while temp != None:
          if count == idx:
              return temp
          else:
              count += 1
              temp = temp.next

  
  # returns the element of the Node at the given index. For invalid idx return None.
  def get(self, idx):
    # To Do
    size = self.countNode()
    if idx < 0 or idx >= size:
      return None
      
    else:
      target = self.nodeAt(idx).element
      return target
      
            
  
  # updates the element of the Node at the given index. 
  # Returns the old element that was replaced. For invalid index return None.
  # parameter: index, element
  def set(self, idx, elem):
    # To Do
    size = self.countNode()
    if idx < 0 or idx >= size:
      return None
      
    else:
      old_elem = self.get(idx)
      count = 0
      self.nodeAt(idx).element = elem
      return old_elem



  # returns the index of the Node containing the given element.
  # if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    count = 0
    temp = self.head

    while temp != None:
        if temp.element == elem:
            temp.element = elem
            return count
        else:
            count += 1
            temp = temp.next

    return -1
  
  # returns true if the element exists in the List, return false otherwise.
  def contains(self, elem):
    # To Do
    count = 0
    temp = self.head

    while temp != None:
        if temp.element == elem:
            return True
        else:
            count += 1
            temp = temp.next

    return False

  # Makes a duplicate copy of the given List. Returns the reference of the duplicate list.
  def copyList(self):
    # To Do
      new_head = None
      new_tail = None
      temp = self.head
      
      while temp != None:
          n1 = Node(temp.element, None)
          if new_head == None:
                  new_head = n1
                  new_tail = n1
                  temp = temp.next
          else:
            new_tail.next = n1
            new_tail = n1
            temp = temp.next
        
      return new_head


  # Makes a reversed copy of the given List. Returns the head reference of the reversed list.
  def reverseList(self):
    # To Do
    reverse_head = None
    temp = self.head

    while temp != None:
        x = temp.next
        temp.next = reverse_head
        reverse_head = temp
        temp = x

    return reverse_head
    
  
  # inserts Node containing the given element at the given index
  # Check validity of index.
  def insert(self, elem, idx):
    # To Do
    size = self.countNode()
    if idx < 0 or idx > size:
      return None

    temp = self.head
    temp1 = self.nodeAt(idx-1)

    if idx == 0:
      new_node = Node(elem, None)
      new_node.next = self.head
      self.head = new_node

    else:
      while temp != None:
        if temp == temp1:
          new_node = Node(elem, temp1.next)
          temp1.next = new_node
          break
        
        else:
          temp = temp.next
   

  # removes Node at the given index. returns element of the removed node.
  # Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    size = self.countNode()
    if idx < 0 or idx >= size:
      return None

    elif idx == 0:
      deleted = self.head
      self.head = self.head.next

    else:
      prev = self.nodeAt(idx - 1)
      deleted = prev.next
      prev.next = deleted.next

    return deleted.element

  
  # Rotates the list to the left by 1 position.
  def rotateLeft(self):
    # To Do
    old_head = self.head
    old_tail = self.tail
    nxt_head = self.nodeAt(self.indexOf(self.head.element)+1)
    
    old_tail.next = self.head
    self.head = nxt_head
    old_head.next = None
    self.tail = old_head
    
  
  # Rotates the list to the right by 1 position.
  def rotateRight(self):
    # To Do
    old_head = self.head
    old_tail = self.tail
    prev_tail = self.nodeAt(self.indexOf(self.tail.element)-1)
    
    old_tail.next = self.head
    prev_tail.next = None
    self.head = old_tail
    self.tail = prev_tail
    
```
