```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  #  Design the constructor based on data type of a. If 'a' is built in python list then
  #  Creates a linked list using the values from the given array.
    self.head = None
    # To Do
    if type(a) == list:
        self.head = Node(None, None, None)
        self.tail = Node(None, None, None)
        
        self.head.next = self.head
        self.head.prev = self.head
        self.tail = self.head

        for i in range(len(a)):
            n = Node(a[i], self.head, self.tail)
            if self.head.element == None:
                self.tail.next = n
                self.head.prev = n
                self.tail = n
            else:
                self.tail.next = n
                self.head.prev = n
                self.tail = n
    
    else:
        self.head = a
  
  # Counts the number of Nodes in the list
  def countNode(self):
    # To Do
    count = 0
    n = self.head.next
    while n != self.head:
        count += 1
        n = n.next
    return count
  
  # prints the elements in the list
  def forwardprint(self):
    # To Do
    n = self.head.next
    while n != self.head:
        print(n.element, end = "")
        n = n.next
        if n != self.head:
            print(", ", end="")
        else:
            print()

  # prints the elements in the list backward
  def backwardprint(self):
    # To Do
    n = self.head.prev
    while n != self.head:
        print(n.element, end = "")
        n = n.prev
        if n != self.head:
            print(", ", end="")
        else:
            print()

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    size = self.countNode()
    if idx < 0 or idx >= size:
      return None
      
    else:
      count = 0
      temp = self.head.next

      while temp != self.head:
          if count == idx:
              return temp
          else:
              count += 1
              temp = temp.next
  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    count = 0
    temp = self.head.next

    while temp != self.head:
        if temp.element == elem:
            temp.element = elem
            return count
        else:
            count += 1
            temp = temp.next

    return -1

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    # To Do
    size = self.countNode()
    if idx < 0 or idx > size:
      return None

    
    if idx == 0:
      new_node = Node(elem, self.head.next, self.head)
      self.head.next.prev = new_node
      self.head.next = new_node

    elif idx == size:
      new_node = Node(elem, self.head, self.tail)
      self.tail.next = new_node
      self.head.prev = new_node
      self.tail = new_node

    else:
      temp = self.head.next
      temp1 = self.nodeAt(idx-1)
      temp2 = temp1.next
      while temp.element != None:
        if temp == temp1:
          new_node = Node(elem, temp2, temp1)
          temp1.next = new_node
          temp2.prev = new_node
          break
        
        else:
          temp = temp.next
   
          
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    size = self.countNode()
    if idx < 0 or idx >= size:
      return None

    elif idx == 0:
      deleted = self.head.next
      self.head.next = self.head.next.next
      self.head.next.prev = self.head

    elif idx == size-1:
        deleted = self.tail
        self.tail = self.tail.prev
        self.tail.next = self.head
        self.head.prev = self.tail

    else:
      prev = self.nodeAt(idx - 1)
      deleted = prev.next
      prev.next = deleted.next
      deleted.next.prev = prev

    return str(deleted.element)
```
